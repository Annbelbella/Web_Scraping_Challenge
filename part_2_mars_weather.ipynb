### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data
table = html_soup.find('table', class_='table')
table_row = table.find_all('tr', class_='data-row')
table_row
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Find the header
table_header = table.find_all('th')

headers=[]

for header_element in table_header:
    
    title=header_element.text
    headers.append(title)
    
headers
```

```python
df=pd.DataFrame(columns=headers)
df
```

```python
# Find the row data 

for row in table.find_all('tr',class_='data-row'):
    
        data = row.find_all('td')
        
        row_data=[td.text.strip() for td in data]
        
        lenghth=len(df)
        df.loc[lenghth]=row_data
        
df
```

```python
# 3. What is the average low temperature by month?

avg_low_temp = df.groupby(['month'])['min_temp'].mean()
avg_low_temp
```

```python
# Plot the average temperature by month

avg_low_temp.plot(kind="bar", color='green')
plt.ylabel("Temperature in Celsius")
plt.xlabel("Months")
plt.title('Avergae Minimum Temperature by Month')
plt.savefig('avg_low_temp.jpeg')
plt.show()
```

```python
# Identify the coldest and hottest months

month_temp = avg_low_temp.sort_values()
month_temp
```

```python
month_temp .plot(kind="bar")
plt.ylabel("Temperature in Celsius")
plt.xlabel("Months")
plt.savefig('sorted_avg_low_temp.jpeg')
plt.show()
```

```python
# 4. Average pressure by Martian month

avg_pressure = df.groupby(['month'])['pressure'].mean()
avg_pressure
```

```python
# Plot the average pressure by month

avg_pressure.sort_values().plot(kind="bar", color='red')
plt.ylabel("Atmospheric Pressure")
plt.xlabel("Month")
plt.title("Average Pressure by month")
plt.savefig('Montly_avg_pressure.jpeg')
plt.show()
```
